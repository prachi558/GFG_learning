### Central limit theorem

```python
#plotting distribution of displacement 
sns.displot(df['displacement'], kind='kde')
```

```python
# creating a list of 30 samples each containing randomly selected 30 elements and taking mean of 
# each element and storing it in a list
```

```python
sample_mean_list = []
for i in range(30):
    sample = df['displacement'].sample(30)
    sample_mean_list.append(sample.mean())
    
```

```python
# plotting sample means distribution
```

```python
sns.displot(sample_mean_list, kind = 'kde')
```

```python
# as we see we get something almost similar to normal distribution
```

```python
# mean of displacemenrt column and mean of normal distribution is similar
import statistics as stat
print(stat.mean(df['displacement']))
print(stat.mean(sample_mean_list))
```

```python
# we did the same thing by increasing number of samples we get and we got a more approximate normal distribution
sample_mean_list = []
for i in range(100):
    sample = df['displacement'].sample(30)
    sample_mean_list.append(sample.mean())
sns.displot(sample_mean_list, kind = 'kde')
print(stat.mean(df['displacement']))
print(stat.mean(sample_mean_list))
```
