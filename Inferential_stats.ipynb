```python
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
## Finding a random sample of 10
```

```python
Random_s = df.sample(10)
Random_s
```

```python
## finding systematic random samples at intervals of 20
```

```python
x = np.arange(0,len(df),20)
df.iloc[x]
```

```python
# stratified random sampling by selecting 10 samples from each subgroup : europe, asia,japan
```

```python
df.groupby('origin').sample(10)
```

```python
df['origin'].value_counts()
```

```python
# since distribution of usa, japan and europe is unequal, instead of taking 10 values of each, take 1% of 
# each subgroup
# above one is disproportionate startified sampling while below one is proportionate stratified sampling
```

```python
df.groupby('origin').sample(frac=0.1)
```
