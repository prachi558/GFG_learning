### generating sine and cos graph

sine wave

```python
sin_x = np.linspace(0, 2*np.pi,50)     # x co-ordinates
sin_y = np.sin(sin_x)                  # y co-ordinates
plt.plot(sin_x,sin_y)                  # plot x and y co-ordinates and connect them
plt.title('sine wave')                 # give title to graph
plt.show()                             # show the plot
```

cosine wave

```python
cos_x = np.linspace((-np.pi),2*np.pi,50)
cos_y = np.cos(cos_x)
plt.plot(cos_x,cos_y)
plt.title('cos wave')
plt.show()
```

sin and cos on same graph with a legend

```python
x = np.linspace((-np.pi),2*np.pi,50)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y, label = 'sine wave')
plt.plot(x,cos_y, label = 'cosine wave')

plt.legend(loc = 'upper right')              #legend creates the index on the graph and label is used to label the graph for 
                                             #the index, loc tells location where legend dhould be

plt.show()
```

using something else to plot graph instead of a solid line

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)

plt.plot(x,y,'*')
plt.show()
```

### saving plots on local machine

```python
plt.style.use('seaborn-dark')

plt.figure(figsize=(18,5))             # 18 is length and 5 is breadth

sin_x = np.linspace(0,2*np.pi,50)
sin_y = np.sin(sin_x)

plt.plot(sin_x,sin_y,linewidth = 5)
plt.title('sine wave')
fig_sin = plt.figure()                # creating a object of the figure to later use it to save the plot
fig_sin.savefig('sine_wave.jpeg')          #sin.png is name of the file saved
```

#### adjusting the plot axes : i.e range for x axis and y axis : xlim for x and ylim for y

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
plt.xlim(1,10)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
plt.ylim(-0.5,0.5)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)
plt.axis([1,10,-0.5,0.5])
plt.plot(x,y)
plt.show()
```

## Histograms and Binnings : data is divided into intervals and bins and height of bar represents points that fall in that bin/ in that interval

```python
d = np.random.randn(100)
plt.hist(d)
plt.show()
```

#### changing number of bins, by default it is 10

```python
plt.hist(d,bins=25)
plt.show()
```

```python
plt.hist(d,bins=30,edgecolor='black',color='red')
plt.show()
```

#### multiple histograms

```python
d1 = np.random.randn(1000)
d2 = np.random.randn(1000) + 2
plt.hist(d1,bins=25,color='red',alpha=0.7)
plt.hist(d2,bins=25,color='blue',alpha=0.7)
plt.show()
```

### 2D Histogram : a regular histogram represents only 1 feature, 2D represents 2

```python
d1 = np.random.randn(1000)
d2 = np.random.randn(1000) + 2
plt.hist2d(d1,d2,bins=25,cmap='viridis')
plt.xlabel('d1')
plt.ylabel('d2')
plt.colorbar()
plt.show()
```

above graph shows the top view showing density of values 

## Pie chart : circular chart divided into slices to represent proportion

```python
food = ['apples','mangoes','strawberries','papaya']
number = [300,240,650,67]
```

```python
plt.pie(number,labels=food)
plt.show()
```

#### cutting certain slices out and putting shadows

```python
explode=[0,0.3,0,0]
plt.pie(number,labels=food,explode=explode,shadow=True)
plt.show()
```

#### adding percentages

```python
plt.pie(number,labels=food,autopct='%1.1f%%')
plt.show()
```

## Multiple subplots

```python
fig,ax = plt.subplots(2,2)    #2,2 represents 2 rows and 2 columns i.e total 4 graphs 
x = np.linspace(1,10)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = x**3

ax[0,0].plot(x,y1)      #at position 0,0
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)
plt.show()
```
